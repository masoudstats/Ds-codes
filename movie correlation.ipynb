```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)

pd.options.display.max_colwidth = 50
```

```python
df = pd.read_csv(r'C:\Users\Latitude 7490\OneDrive\Desktop\movies.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.dropna(axis=0, inplace=True)
df.shape
```

```python
df.released
```

```python
index = df.released.astype(str).str.find(' (')
index
```

```python
pd.set_option('mode.chained_assignment', None)

df['core_year'] = df.released

for idx, val in  df.released.items():
    df.core_year[idx] = val[index[idx]-4: index[idx]]
    
df.core_year = df.core_year.astype('int64')
df.core_year
```

```python
df
```

```python
df.dtypes
```

```python
df.gross = df.gross.astype('int64')
df.budget = df.budget.astype('int64')
```

```python
df.sort_values('gross', inplace=True, ascending=False)
df.gross
```

```python
plt.scatter(df.budget, df.gross)
```

```python
sns.regplot(x=df.budget, y=df.gross, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
corr_matrix = df.corr(method='pearson')
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True)
```

```python
df_numerized = df

for col in df_numerized.columns:
    if(df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
        
df_numerized    
```

```python
corr_matrix = df_numerized.corr(method='pearson')
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True)
```

```python
unstack_corr = corr_matrix.unstack()
unstack_corr
```

```python
sorted_corr = unstack_corr.sort_values()
sorted_corr
```

```python
high_corr = sorted_corr[sorted_corr >= 0.5]
high_corr
```
